```python
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name=lora_model_name,
    max_seq_length=2048,
    dtype=None,
    load_in_4bit=True
)
```

```python
FastLanguageModel.for_inference(model)
```

```python
dataset = load_dataset("knkarthick/dialogsum")
```

```python
test_data_size = 30
```

```python
test_dataset = dataset['test'].shuffle(seed=42).select(range(test_data_size))
```

```python
test_dialogues = [sample['dialogue'] for sample in test_dataset]
test_summaries = [sample['summary'] for sample in test_dataset]
```

```python
alpaca_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
{}

### Input:
{}

### Response:
{}"""
```

```python
instruction='Write a concise dialogue of the following summary.'
```

```python
predicted_summaries = []
```

```python
for gold_dialogue in tqdm(test_summaries):

    try:
        prompt = alpaca_prompt.format(
            instruction,
            gold_dialogue,
            ""
        )

        inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

        outputs = model.generate(
            **inputs,
            max_new_tokens=1280,
            temperature=0,
            use_cache=True,
            pad_token_id=tokenizer.eos_token_id
        )

        prediction = tokenizer.decode(
            outputs[0][inputs.input_ids.shape[-1]:],
            skip_special_tokens=True,
            cleanup_tokenization_spaces=True
        )

        predicted_summaries.append(prediction)

    except Exception as e:
        print(e) # log error and continue
        continue
```

```python
predicted_summaries[1]
```

```python
bert_scorer = evaluate.load("bertscore")
```

```python
score = bert_scorer.compute(
    predictions=predicted_summaries,
    references=test_dialogues,
    lang='en',
    rescale_with_baseline=True
)
```

```python
sum(score['f1'])/len(score['f1'])
```
